```python
import stanza
stanza.download('hi')
```

###### Creating a NLP Pipeline

```python
nlp = stanza.Pipeline(lang='hi', processors='tokenize, pos, lemma, depparse')
```

*About the dataset :*

The Jātaka tales are a voluminous body of literature native to India concerning the previous births of Gautama Buddha in both human and animal form. The future Buddha may appear as a king, an outcast, a god, an elephant—but, in whatever form, he exhibits some virtue that the tale thereby inculcates. Often, Jātaka tales include an extensive cast of characters who interact and get into various kinds of trouble - whereupon the Buddha character intervenes to resolve all the problems and bring about a happy ending.

```python
with open ("F:/Dataset/Jaatak_kathaye.txt" , 'r',encoding='utf16', errors = 'strict') as f:
    txt = f.read()

print(txt)
```

```python
doc = nlp(txt)
```

```python
HINDI_TOKENS = []
for i, sentence in enumerate(doc.sentences):
    sentence_tokens = []
    [sentence_tokens.append(token.text) for token in sentence.tokens]
    HINDI_TOKENS.append(sentence_tokens)
```

```python
print("No of tokens: ", len(HINDI_TOKENS))
for sent_tok in HINDI_TOKENS:
    print(f'{sent_tok}')
```

```python
hindi_pos_tags = [(word.text, word.upos, word.xpos) for sent in doc.sentences for word in sent.words]
print(hindi_pos_tags)
```

```python
hindi_words_lemma = [(word.text, word.lemma) for sent in doc.sentences for word in sent.words]
print(hindi_words_lemma)
```

```python
hindi_words_feats = [(word.text, word.feats) for sent in doc.sentences for word in sent.words]
print(hindi_words_feats)
```

```python
print(*[f'id: {word.id}\tword: {word.text}\thead id: {word.head}\thead: {sent.words[word.head-1].text if word.head > 0 else "root"}\tdeprel: {word.deprel}' for sent in doc.sentences for word in sent.words], sep='\n')
```
